```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Projetos_codecademy/WHO_project/all_data.csv')
df.head()
```

```python
years=df.Year
country=df.Country
df_chile=df[df['Country']=='Chile']
df_china=df[df['Country']=='China']
df_germany=df[df['Country']=='Germany']
df_Mexico = df[df['Country']=='Mexico']
df_USA = df[df['Country']=='United States of America']
df_Zimbabwe = df[df['Country']=='Zimbabwe']
```

```python
figure=plt.figure(figsize=(10,10))
plt.subplot(3,3,1)
plt.scatter(df_Zimbabwe['Life expectancy at birth (years)'],df_Zimbabwe.GDP,color='black',label='Zimbabwe',s=15)
plt.legend(loc='upper left')
plt.ylabel('GDP in Trillions of US Dollars')
plt.subplot(3,3,2)
plt.scatter(df_china['Life expectancy at birth (years)'],df_china.GDP,color='red',label='China',s=15)
plt.legend(loc='upper left')
plt.subplot(3,3,3)
plt.scatter(df_germany['Life expectancy at birth (years)'],df_germany.GDP,color='orange',label='Germany',s=15)
plt.legend(loc='upper left')
plt.subplot(3,3,4)
plt.scatter(df_Mexico['Life expectancy at birth (years)'],df_Mexico.GDP,color='green',label='Mexico',s=15)
plt.legend(loc='upper left')
plt.ylabel('GDP in Trillions of US Dollars')
plt.xlabel('Life Expectancy')
plt.subplot(3,3,5)
plt.scatter(df_USA['Life expectancy at birth (years)'],df_USA.GDP,color='blue',label='USA',s=15)
plt.legend(loc='upper left')
plt.xlabel('Life Expectancy')
plt.subplot(3,3,6)
plt.scatter(df_chile['Life expectancy at birth (years)'],df_chile.GDP,color='brown',label='Chile',s=15)
plt.legend(loc='upper left')
plt.xlabel('Life Expectancy')
plt.subplots_adjust(wspace=0.35,hspace=0.35)
plt.show()
figure.savefig('Scatter_between_GDP_and_LE.png')
```

In this visualization, it becomes evident that exists a generally positive correlation between Life Expectancy and GDP across all six countries. Certain nations display a more linear correlations, such as the United States and Mexico. China's plot, suggests a slightly more exponential curve in comparison.

```python
plt.figure(figsize=(10,8))
plt.plot(df_chile.Year,df_chile.GDP,color='brown',label='Chile')
plt.plot(df_china.Year,df_china.GDP,color='red',label='China')
plt.plot(df_germany.Year,df_germany.GDP,color='orange',label='Germany')
plt.plot(df_Mexico.Year,df_Mexico.GDP,color='green',label='Mexico')
plt.plot(df_USA.Year,df_USA.GDP,color='blue',label='USA')
plt.plot(df_Zimbabwe.Year,df_Zimbabwe.GDP,color='black',label='Zimbabwe')
plt.xlabel('Years')
plt.ylabel('GDB in Trillions of US Dollars')
plt.legend()
figure.savefig('Line_between_GDP_and_Years.png')
plt.show()
```

In this visualization, it's apparent that the USA exhibits a linear growth pattern. On the other hand, China demonstrates a slightly exponential growth, similar to the first plot. The remaining countries appear to have experienced minimal GDP growth between 2000 and 2015. 

```python
plt.figure(figsize=(12,10))
plt.subplot(3,3,1)
plt.plot(df_Zimbabwe.Year,df_Zimbabwe.GDP,color='black',label='Zimbabwe')
plt.legend()
plt.ylabel('GDB in Trillions of US Dollars')
plt.subplot(3,3,2)
plt.plot(df_china.Year,df_china.GDP,color='red',label='China')
plt.legend()
plt.subplot(3,3,3)
plt.plot(df_germany.Year,df_germany.GDP,color='orange',label='Germany')
plt.legend(loc='upper left')
plt.subplot(3,3,4)
plt.plot(df_Mexico.Year,df_Mexico.GDP,color='green',label='Mexico')
plt.legend()
plt.xlabel('Years')
plt.ylabel('GDB in Trillions of US Dollars')
plt.legend()
plt.subplot(3,3,5)
plt.plot(df_USA.Year,df_USA.GDP,color='blue',label='USA')
plt.xlabel('Years')
plt.legend()
plt.subplot(3,3,6)
plt.plot(df_chile.Year,df_chile.GDP,color='brown',label='Chile')
plt.xlabel('Years')
plt.legend()
plt.subplots_adjust(wspace=0.35,hspace=0.35)
figure.savefig('Line_between_GDP_and_Years_each_country.png')
plt.show()
```

 Now, with a clearer visualization, we can observe the GDP growth trends from 2000 to 2015 across the six countries. It's evident that all nations experienced GDP growth during this period, the biggests ones being, Zimbabwe, Chile and China.
